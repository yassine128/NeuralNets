```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import neuralNets as NN
from sklearn.datasets import make_moons
```

```python
# Generate moon-shaped data
num_points = 200
X, y = make_moons(n_samples=num_points, noise=0.2, random_state=42)

# Visualize the generated data
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Moon-Shaped Data')
plt.show()
```

```python
mlp = NN.MLP([2, 5, 1], NN.Activation().sigmoid(), NN.Activation().sigmoid_derivative())
loss_values = mlp.train(X, y, 0.1, 700)
```

```python
# Extract loss values from the array structure
loss_values = [item[0, 0] for item in loss_values]

# Plot the loss graph
plt.figure(figsize=(10, 5))
plt.plot(range(len(loss_values)), loss_values, color='blue')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Graph')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from matplotlib.animation import FuncAnimation

# Define the number of epochs
num_epochs = 400

# Initialize the MLP and loss_values list
mlp = NN.MLP([2, 5, 5, 1], NN.Activation().sigmoid(), NN.Activation().sigmoid_derivative())  # Define your MLP architecture
loss_values = []

# Create a figure and axis for plotting
fig, ax = plt.subplots()

# Define a function to update the plot for each epoch
def update(epoch):
    ax.clear()

    # Train the model for this epoch
    loss = mlp.train(X, y, learning_rate=0.1, epochs=1)
    loss_values.append(loss)

    # Plot the decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    predictions = mlp.forward(grid_points)
    predictions = predictions.reshape(xx.shape)
    ax.contourf(xx, yy, predictions, cmap=plt.cm.Spectral, alpha=0.2)

    # Plot the training data
    ax.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
    ax.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')

    # Plot settings
    ax.legend()
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(f'Epoch {epoch+1}/{num_epochs}')

# Create the animation
animation = FuncAnimation(fig, update, frames=num_epochs, repeat=False)

# Save the animation as a GIF
animation.save('epoch_animation.gif', writer='imagemagick')
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate circle-shaped data
num_points = 230
X, y = make_circles(n_samples=num_points, noise=0.1, factor=0.2, random_state=42)

# Separate the two classes
class_1_indices = y == 0
class_2_indices = y == 1

# Visualize the generated data
plt.scatter(X[class_1_indices][:, 0], X[class_1_indices][:, 1], color='red', label='Class 1')
plt.scatter(X[class_2_indices][:, 0], X[class_2_indices][:, 1], color='blue', label='Class 2')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Circle-Shaped Data')
plt.show()
```
